```python
file = load_workbook(filename = 'cafe-sby.xlsx')
data = file['cafe - Copy'] #sheet pada excel

df = pd.DataFrame(data.values)
df
```

```python
# mendefinisikan dokumen dari masing-masing kolom

df.columns = ['DOC', 'TEXT']
df
```

```python
# mengambil salah satu kolom dan ditampung pada variabel isi_text

isi_text = df[['TEXT']]
isi_text
```

```python
# mengubah data ke dalam bentuk list menggunakan values.tolist

data = []
data = isi_text['TEXT'].values.tolist()
data
```

```python
# menghapus judul TEXT
data.remove('TEXT')
```

### Proses Preprocessing

```python
#merubah huruf di dalam data menjadi kecil
case_folding = [] # deklarasikan variabel case_folding yang merupakan variabel list ynag berfungsi untuk menampung 
                    # hasil proses lower case

for x in data:
    kecil = x.lower()
    case_folding.append(kecil)
print(case_folding)
```

```python
# import method word_tokenize dari library nltk.tokenize
from nltk.tokenize import word_tokenize

# tokenize

tokenisasi = []

for x in case_folding:
    proses_tokenisasi = word_tokenize(x)
    tokenisasi.append(proses_tokenisasi)
print(tokenisasi)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('Indonesian'))
```

```python
# filterisasi dari kata-kata yang ada di dalam stopwords
filterisasi = []

i = 0
for x in tokenisasi:
    filterisasi.append([])
    for y in x:
        if y not in stop_words:
            filterisasi[i].append(y)
    i+=1
```

```python
# proses penghapusan tanda baca, angka
cleaning = []

i = 0

for x in filterisasi:
    cleaning.append([])
    for y in x:
        if y.isalpha():
            cleaning[i].append(y)
    i+=1
```

```python
# membandingkan data dalam filterisasi dan cleaning
import numpy as np
np.shape(filterisasi[0]), np.shape(filterisasi[1]), np.shape(filterisasi[2]), np.shape(filterisasi[3]), np.shape(filterisasi[4]), 
```

```python
np.shape(cleaning[0]), np.shape(cleaning[1]), np.shape(cleaning[2]), np.shape(cleaning[3]), np.shape(cleaning[4]), 
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
# fungsi stemming untuk menghilangkan imbuhan aalan dan akhiran
factory = StemmerFactory()
stemmer = factory.create_stemmer()
stemming = []
i = 0

for x in cleaning:
    stemming.append([])
    for y in cleaning[i]:
        stemming[i].append(stemmer.stem(y))
    i+=1
```

```python
data_bersih = []
i = 0

for x in stemming:
    data_bersih.append((' '.join(map(str, x)))) #proses penggabungan kalimat dari data yang telah di stemming
    i+=0

data_bersih
```

### Proses Penghitungan Pembobotan Kata

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
```

```python
CV = CountVectorizer() #melakukan pemanggilan method
term_fit = CV.fit(data_bersih)

print(len(term_fit.vocabulary_)) #untuk mengetahui beberapa kata yang ada di dalam dokumen
```

```python
print(term_fit.vocabulary_)

# disini terdapat beberapa kata yang tersimpan di dalam kamus term_fit
# tampilan data ada berupa kata beserta value atau nilai dari persebaran kata-kata yang terdapat didalam dokumen
```

```python
# proses mencari frekuensi pada kata - kata yang terdapat di dalam dokumen

frekuensi_semua_kata = term_fit.transform(data_bersih)
print(frekuensi_semua_kata)
```

```python
# tf-idf atau bobot dari kata kata yang terdapat pada dokumen

dokumen = term_fit.transform(data_bersih)
tfidf_transformer = TfidfTransformer().fit(dokumen)
print(tfidf_transformer.idf_)

tfidf = tfidf_transformer.transform(frekuensi_semua_kata)
print(tfidf)
```
